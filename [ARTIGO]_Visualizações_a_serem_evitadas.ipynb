Este artigo tem o objetivo de apresentar algumas formas de visualização de dados que são extremamente comuns em grandes empresas, jornais, televisão e artigos científicos, e que de certa maneira, são visualizações ruins.

## Gráfico de Pizza

<center> <img width='700' src='https://image.freepik.com/free-photo/relaxing-girls-watching-movie-eating-pizza_23-2147764743.jpg'>

Amo pizza, mas não utilizaria esse gráfico em um relatório. Resumindo, gráfico de pizza são ruins!

Para entender melhor, vamos ver um exemplo.

O gráfico abaixo representa a participação no mercado de quatro fornecedores. Faça uma análise rápida - qual é o maior fornecedor desse mercado?

gráfico.png

Rapidamente você conseguiu identificar o maior fornecedor? Se você precisasse avaliar a proporção que o Fornecedor B ocupa no mercado, que porcentagem você poderia estimar?

- 35%?
- 40%?

Veja que acontece quando achatamos e adicionamos os números às fatias.

```python
#@title
# Importando bibliotecas
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# variáveis
dados = [34,31,26,9]
cores = ('mediumblue','royalblue','dodgerblue','blue')
labels = ['Fornecedor A','Fornecedor B','Fornecedor C','Fornecedor D']

# Criando o gráfico de pizza e configurações iniciais
plt.pie( dados,
        explode=[0.05,0.05,0.05,0.05], 
        colors=cores,
        shadow=True, 
        autopct='%1.1f%%',
        counterclock=False,
        startangle=120,
        )

# Adicionando título e legenda
plt.title("Participação no mercado de fornecedores", 
          fontsize=16, 
          color='gray',
          weight='bold')

plt.legend(loc="right", labels=labels,bbox_to_anchor=(1, 0, 0.5, 1))

# Finalizando processo
plt.show()
```

Muitas pessoas diriam que o Fornecedor B era o maior, e quando adicionamos os valores foi possível observar que o Fornecedor A era o maior. Mas por quê isso acontece?

Existem dois problemas que apresentam um desafio para a interpretação correta desses dados. O primeiro deles é a utilização do gráfico de pizza em 3D. Gráficos 3D não servem para nada e podem distorcer a percepção visual dos números, nesse caso, ele torna as partes superiores mais afastados e menores do que realmente são, enquanto as partes inferiores ficam mais próximas e maiores. Diante disso, nossa primeira regra: Não use gráficos 3D!

Quando tiramos o 3D o gráfico fica achatado, e aí vem o segundo problema e o porquê dos gráficos de pizza serem ruins. O olho humano não consegue atribuir corretamente valores quantitativos no espaço bidimensional, ou seja, é difícil ler gráficos de pizza. Quando os segmentos têm tamanhos parecidos é difícil (às vezes impossível) dizer qual é o maior. E quando identificamos qual é o maior, não é possível avaliar quanto.

```python
#@title
# Importando bibliotecas
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

dados = [34,31,26,9]
cores = ('mediumblue','royalblue','dodgerblue','blue')
labels = ['Fornecedor A','Fornecedor B','Fornecedor C','Fornecedor D']

# Criando uma fig
fig, ax = plt.subplots(figsize=(10, 5))

#Configurando gráfico em formato de rosca
ax.pie(dados, 
        colors=cores,
        shadow=True, 
        counterclock=False,
        startangle=120,
       )

# Adicionando título e legenda
plt.title("Participação no mercado de fornecedores", 
          fontsize=16, 
          color='gray',
          weight='bold')

plt.legend(loc="right", labels=labels,bbox_to_anchor=(1, 0, 0.5, 1))

# Finalizando processo
plt.show()
```

Para solucionar o problema, poderia adicionar números dentro do gráfico ou inserir legendas, mas o visual não vale o espaço que ocupa e aumentaria a carga cognitiva desnecessariamente.

Se você for utilizar um gráfico de pizza, pare e pergunte-se: por quê? Se puder responder essa pergunta, possivelmente terá refletido suficientemente para utilizá-la, mas certamente esse não deve ser o primeiro tipo de gráfico a ser empregado, dadas as dificuldades citadas. 

Se é difícil de ler, vale a pena?

## Gráfico de Rosca ou Donuts

<center> <img width='500' src='https://image.freepik.com/free-photo/delicious-donuts_144627-6267.jpg'>

Após morar alguns meses no Canadá, criei um novo amor além da pizza: os Donuts! Apesar de serem macios e deliciosos (provavelmente você gostaria de um agora), não os utilize para visualização de dados!

```python
#@title
# Criando uma fig
fig, ax = plt.subplots(figsize=(10, 5))

#Configurando gráfico em formato de rosca
ax.pie(dados, 
        colors=cores,
        shadow=True, 
        counterclock=False,
        startangle=120,
        wedgeprops=dict(width=0.5)
       )
       
# Adicionando título e legenda
plt.title("Participação no mercado de fornecedores", 
          fontsize=16, 
          color='gray',
          weight='bold')

plt.legend(loc="right", labels=labels,bbox_to_anchor=(1, 0, 0.5, 1))

# Finalizando processo
plt.show()
```

Com a visualização em formato de pizza, pedimos ao público-alvo para comparar ângulos e áreas. Com um gráfico de donuts, ou rosca, estamos pedindo para comparar comprimento de arcos. Quão confiante você se sente para atribuir valores quantitativos a um comprimento de arco?

Não muito? Pois bem. Não use gráficos de rosca!



## Solução para gráficos de pizza e rosca

Ok! Gráficos de pizza e rosca são ruins. Então, qual é a solução?

Podemos simplesmente utilizar gráficos de barras horizontais!

```python
#@title
fig, ax = plt.subplots(figsize=(8,5))

ax.barh(labels,dados,color=('mediumblue'))
      
#Removendo bordas
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.spines["bottom"].set_visible(False)

# Removendo e ajustando ticks
plt.yticks(fontsize=12)

plt.tick_params(
    axis='both',         
    which='both',     
    bottom=False,
    top=False,
    left=False,        
    labelbottom=False,
    ) 

# Ordenando do maior para o menor
ax.invert_yaxis()

# Adicionando título e legenda
ax.text(-6.5,-1,"Participação no mercado de fornecedores",
          fontsize=16, 
          color='gray',
          weight='bold',
          horizontalalignment= 'left')
          
# Inserindo porcentagem nas barras
ax.text(34-3, 0, "34%", fontsize=12, color='white',weight='bold')
ax.text(32-4, 1, "32%", fontsize=12, color='white',weight='bold')
ax.text(26-3, 2, "26%", fontsize=12, color='white',weight='bold')
ax.text(9-2, 3, "9%", fontsize=12, color='white',weight='bold')

# Finalizando processo
fig.tight_layout()
plt.show()
```

Com gráficos de barras, nossos olhos comparam os pontos extremos. Como as barras são alinhadas em uma linha de base comum, é fácil avaliar o tamanho relativo. Isso facilita ver não apenas qual segmento é maior, mas também o quanto, incrementalmente, é maior que os outros segmentos.

Diferentemente do gráfico de pizza, o gráfico de barras permite ordenar a visualização de acordo com alguma ordem natural ou prioridade que queira ser apresentada.

## Gráficos  de Espaguete

<center><img width='700' src='https://image.freepik.com/free-photo/family-four-eating-spaghetti_1398-5031.jpg'>

Certeza que você já está com água na boca ansioso pelo almoço, e nesse almoço cairia muito bem um espaguete, porém para a sua apresentação, evite esse gráfico. Gráfico de espaguete é uma visualização no qual as linhas se sobrepõem, tornando difícil focar uma série de dados por vez.

```python
#@title
# Importando bibliotecas
import numpy as np
from numpy import random

# Criando valores
x=np.arange(2015,2021)
y1=random.randint(101, size=(6))
y2=random.randint(101, size=(6))
y3=random.randint(101, size=(6))
y4=random.randint(101, size=(6))
```

```python
#@title
# Importando bibliotecas
import matplotlib.ticker as mtick 
from matplotlib.ticker import PercentFormatter

# Criando Fig
fig, ax = plt.subplots(figsize=(8,5))

# Criando linhas
ax.plot(x,y2, linewidth=2)
ax.plot(x,y3,linewidth=2)
ax.plot(x,y4, linewidth=2)
ax.plot(x,y1, linewidth=2)

# Removendo bordas
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)

# Ajustando ticks
plt.tick_params(
    axis='both',         
    which='both',     
    bottom=False,
    top=False,
    left=False,        
    labelbottom=True,
    ) 

# Ajustando eixo Y com formato de porcentagem
ax.yaxis.set_major_formatter(mtick.PercentFormatter()) 

# Criando Tíítulo
ax.text(2014.33,110,"Diponibilidade média das aeronaves",
          fontsize=16, 
          color='gray',
          weight='bold',
          horizontalalignment= 'left')


# Criando legenda
plt.legend(labels=("C-105","C-98","H-60L","F-5"),bbox_to_anchor=(1,1))

# Finalizando processo
plt.show()
```

Esse tipo de gráfico é tão confuso quanto jogar um punhado de talharim cru no chão. Perceba a dificuldade que é focar em uma única linha. Esse fato ocorre por elas estarem competindo por sua atenção.

Um forma de simples de resolver o problema do espaguete, seria usar atributos pré-atentivos a fim de chamar a atenção para uma linha por vez. Veja o exemplo:

```python
#@title
# Criando Fig
fig, ax = plt.subplots(figsize=(10,5))

# Criando linhas
ax.plot(x,y2, linewidth=2,color='gainsboro')
ax.plot(x,y3,linewidth=2,color='gainsboro')
ax.plot(x,y4, linewidth=2,color='gainsboro')
ax.plot(x,y1, linewidth=2.5,color='red')

# Removendo bordas
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)

# Ajustando ticks
plt.tick_params(
    axis='both',         
    which='both',     
    bottom=False,
    top=False,
    left=False,        
    labelbottom=True,
    ) 

# Ajustando eixo Y com formato de porcentagem
ax.yaxis.set_major_formatter(mtick.PercentFormatter()) 

# Inserindo texto
ax.text(x[5]+0.1, y1[5], "F-5", fontsize=14,weight='bold',color='brown')

# Criando tíítulo
ax.text(2014.33,110,'Diponibilidade média das aeronaves',
          fontsize=16, 
          color='gray',
          weight='bold',
          horizontalalignment= 'left')

# Finalizando processo
plt.show()
```

Ou, poderíamos desembaraçar o gráfico separando as linhas espacialmente, como a seguir:

```python
#@title
# Criando Fig
fig, (ax1,ax2,ax3,ax4) = plt.subplots(4,1,figsize=(10,8),sharex=True)

# Criando linhas
ax1.plot(x,y1, linewidth=2, color= 'r')
ax2.plot(x,y2,linewidth=2, color= 'g')
ax3.plot(x,y3, linewidth=2, color='orange')
ax4.plot(x,y4, linewidth=2, color='b')

#Removendo bordas
ax1.spines["bottom"].set_visible(False)
ax1.spines["right"].set_visible(False)
ax1.spines["left"].set_visible(False)
ax2.spines["right"].set_visible(False)
ax2.spines["left"].set_visible(False)
ax3.spines["right"].set_visible(False)
ax3.spines["left"].set_visible(False)
ax4.spines["right"].set_visible(False)
ax4.spines["left"].set_visible(False)
ax2.spines["bottom"].set_visible(False)
ax3.spines["bottom"].set_visible(False)

# ajustando ticks
ax1.tick_params(
    axis='both',         
    which='both',     
    bottom=False,
    top=False,
    left=False,        
    labelbottom=False,
    labelleft=False
    ) 
ax2.tick_params(
    axis='both',         
    which='both',     
    bottom=False,
    top=False,
    left=False,        
    labelbottom=False,
    labelleft=False
    ) 
ax3.tick_params(
    axis='both',         
    which='both',     
    bottom=False,
    top=False,
    left=False,        
    labelbottom=False,
    labelleft=False
    ) 
ax4.tick_params(
    axis='both',         
    which='both',     
    bottom=False,
    top=False,
    left=False,        
    labelbottom=True,
    labelleft=False
    ) 

# Criando label para o eixo y
ax1.set_ylabel("F-5", size=14,rotation=0, labelpad = 70, weight='bold', color='gray')
ax2.set_ylabel("H-60L", size=14,rotation=0, labelpad = 70, weight='bold',color='gray')
ax3.set_ylabel("C-98", size=14,rotation=0, labelpad = 70, weight='bold',color='gray')
ax4.set_ylabel("C-105", size=14,rotation=0, labelpad = 70, weight='bold',color='gray')

# Craindo texto numérico
ax1.text(x[0]-0.3, y1[0]-10, str(y1[0])+"%", fontsize=14, weight='bold',color='r')
ax2.text(x[0]-0.3, y2[0]-10, str(y2[0])+"%", fontsize=14, weight='bold',color='g')
ax3.text(x[0]-0.3, y3[0]-10, str(y3[0])+"%", fontsize=14, weight='bold',color='orange')
ax4.text(x[0]-0.3, y4[0]-10, str(y4[0])+"%", fontsize=14, weight='bold',color='b')

ax1.text(x[5]+0.1, y1[5]-10, str(y1[5])+"%", fontsize=14, weight='bold',color='r')
ax2.text(x[5]+0.1, y2[5], str(y2[5])+"%", fontsize=14, weight='bold',color='g')
ax3.text(x[5]+0.1, y3[5]-10, str(y3[5])+"%", fontsize=14, weight='bold',color='orange')
ax4.text(x[5]+0.1, y4[5]-10, str(y4[5])+"%", fontsize=14, weight='bold',color='b')

# Criando tíítulo
ax1.text(2017.5,130,'Diponibilidade média das aeronaves',
          fontsize=16, 
          horizontalalignment= 'center')

# Finalizando processo
plt.show()
```

Não existe uma única resposta certa, cada situação tem suas peculiaridades e melhores soluções, o importante é: Quando se deparar com um gráfico de espaguete saiba que existem formas de visualizações mais eficientes.

## Gráfico de Waffle

<center><img width='700' src='https://image.freepik.com/free-photo/waffle_95678-13.jpg'>

Para finalizar, vamos para a sobremesa!

Apesar de ser uma excelente sobremesa com sorvete, tome muito cuidado com esse tipo de visualização. Assim como nos gráficos 3D, o gráfico de waffle apresenta dificuldades de interpretação pelo os olhos humanos por ser um gráfico bidimensional.

Uma forma de utilizá-lo, seria em forma de área com números de magnitudes amplamente diferentes. Veja o exemplo:

```python
#@title
# instalar o pywaffle
!pip install pywaffle -q
```

```python
#@title
# importar as bibliotecas necessárias
from pywaffle import Waffle

# exemplo de Waffle
fig = plt.figure(
                FigureClass=Waffle,
                rows=10,
                columns=10,
                vertical=True,
                starting_location='NW',
                title={
                'label': 'Curso de Data Science',
                'loc': 'left',
                'fontdict': {
                'fontsize': 19.5, 
                }}, 
                values={'35 Candidatos Inscritos': 35,
                        '16 Alunos Selecionados': 16,
                        '9 Alunos Aprovados': 9},
                legend={'loc': 'upper left',
                        'bbox_to_anchor': (1.1, 1)}
                )

fig.set_tight_layout(False)
plt.show()
```

Esse tipo de visualização, utilizado de forma correta, pode ser um diferencial na análise de desempenho e no controle de metas, funcionando muito bem como um fator psicológico motivador de uma forma mais descontraída e informal.

```python
#@title
import pandas as pd
data = pd.DataFrame(
    {
        'labels': ['Marcelo','Luciana','Vera'],
        'Maio': [27, 35, 56],
        'Junho': [42, 61, 28],
        'Julho': [55, 42, 37],
    },
).set_index('labels')
```

```python
#@title
fig = plt.figure(
    FigureClass=Waffle,
    plots={
        '311': {
            'values':data['Maio'],
            'labels':[f"{k} ({v})" for k, v in data['Maio'].items()],
            'legend': {
                'loc': 'upper left',
                'bbox_to_anchor': (1.05, 1),
                'fontsize': 10
            },
            'title': {
                'label': 'Vendas Maio',
                'loc': 'left',
                'fontsize': 14
            }
    },
    
        '312': {
            'values':data['Junho'],
            'labels':[f"{k} ({v})" for k, v in data['Junho'].items()],
            'legend': {
                'loc': 'upper left',
                'bbox_to_anchor': (1.05, 1),
                'fontsize': 10
            
        },
        'title': {
                'label': 'Vendas Junho',
                'loc': 'left',
                'fontsize': 14
            }
        },

        '313': {
            'values':data['Julho'],
            'labels':[f"{k} ({v})" for k, v in data['Julho'].items()],
            'legend': {
                'loc': 'upper left',
                'bbox_to_anchor': (1.05, 1),
                'fontsize': 10
        },
        'title': {
                'label': 'Vendas Julho',
                'loc': 'left',
                'fontsize': 14
            }
    }
  },
    rows=5,
    icons='tshirt',
     colors=["#232066", "#983D3D", "#DCB732"],
    font_size=12,
    figsize=(9,6))



tight_Layout=False
plt.show()
```
